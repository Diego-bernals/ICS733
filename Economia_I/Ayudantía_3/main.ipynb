# 1° Ejercicio

La firma que produce aviones de papel tiene la siguiente función de producción siendo K (maquinas) y L (colboradores).

$$
F(W,L) = 5L^{0.5} + 2K^{0.6}
$$

- Comente cuales son lass relaciones entre costo total, costo marginal y costo mdio en el corto y largo plazo.

**Respuesta**: 

En el corto plazo, la firma que produce aviones de papel tiene restringida su producción por algún factor —en la mayoría de los casos, el capital—. En cambio, en el largo plazo, la firma tiene libre elección de los factores que minimizan el costo. Por lo tanto, los costos totales son mayores en el corto plazo que en el largo plazo.

Luego, los costos medios se definen como los costos totales promedio. Si comparamos estos costos en el corto y largo plazo, observamos lo siguiente: en el gráfico, los costos medios de largo plazo están por debajo de los de corto plazo, excepto en el punto donde la cantidad producida es óptima, donde ambos coinciden.

En cambio, los costos marginales se comportan distinto. En el gráfico, los costos marginales de corto plazo son menores que los de largo plazo antes de llegar a la cantidad óptima, pero después de ese punto los costos marginales de corto plazo se vuelven mayores que los de largo plazo.

- Supongamos que $K=6$ en el corto plazo, ¿Cual es el producto marginal del trabajo?. Calcule la función de costo total en el corto plazo.

**Respuesta**: 

Si $K = 6$ en el corto plazo esto implica que la función de producción quedaria de la siguiente manera

$$
F(6,L) = 5L^{0.5} + 2 \cdot 6^{0.6}
$$

El producto marginal en relación a el trabajo quedaria:

$$
PMG_{L} = \frac{dq}{dL} = \frac{2.5}{\sqrt{L}}
$$

El producto medio seria la producción total dividida en la cantidad de trabajo (L):

$$
PME_{L} = \frac{5L^{0.5} + 2\cdot6^{0.6}}{L}
$$

$$
PME_{L} = \frac{5}{\sqrt{L}} + \frac{2\cdot6^{0.6}}{L}
$$


Función de costos totales:

Primero dejamos expresado en terminos de *q* la función de produccion de corto plazo.
$$
q = 5L^{0.5} + 2 \cdot 6^{0.6}
$$
$$
\frac{q - 2 \cdot 6^{0.6}}{5} = L^{0.6}
$$
$$
L = (\frac{q - 2 \cdot 6^{0.6}}{5})^{2}
$$

La función de costos totales $C = wL + rK$, reemplazando:

$$
C = w \cdot (\frac{q - 2 \cdot 6^{0.6}}{5})^{2} + 6r
$$

- Tiene la función de producción rendimientos crecientes, decrecientes o constantes a escala?, ¿ Cual es la elasticidad de sustitución entre capital (K) y trabajo (L).

**Respuesta:**

Multiplicamos ambas variables por una escalar $t$ y evaluamos:

$$
F(tW,tL) = 5(tL)^{0.5} + 2(tK)^{0.6}
$$

$$
F(tW,tL) = 5(t)^{0.5}\cdot(L)^{0.5} + 2(t)^{0.6}\cdot(K)^{0.6}
$$

Al no poder factorizar por un $t^{alpha}$ comun no podemos afirmar que tenemos una proporción constante a escala, lo que implica que no podemos afirmar si tenemos una función creciente o decreciente :c.

- parte 2

$$
F(K, L) = 5L^{0.5} + 2K^{0.6}
$$

Denotaremos la elasticidad de la sustitución como $\sigma$

$$
\sigma = \frac{d\ln(K/L)}{d\ln(TMST)}
$$

Derivamos con respecto a cada factor:

Producto marginal del trabajo
$$
PMg_L = \frac{dF}{dL} = \frac{5}{2} L^{-0.5}
$$

porcusto marginal del capital
$$
PMg_K = \frac{dF}{dK} = 2 \cdot 0.6 K^{-0.4} = 1.2 K^{-0.4}
$$

Reemplazando en $\sigma$:

$$
TMST = \frac{PMg_L}{PMg_K} = \frac{\frac{5}{2} L^{-0.5}}{1.2 K^{-0.4}} = \frac{25}{12} \cdot \frac{K^{0.4}}{L^{0.5}}
$$

Aplicamos el *ln*:

$$
\ln(TMST) = \ln\left(\frac{25}{12}\right) + 0.4 \ln K - 0.5 \ln L
= \ln\left(\frac{25}{12}\right) + 0.4 \ln\left(\frac{K}{L}\right) + 0.1 \ln L
$$

derivamos con respecto a $ln(\frac{K}{L})$
$$
\sigma = 2.5
$$

$$
\therefore \text{Al ser una elasticidad positiva podemos concluir que hay una alta sustitución entre Capital y Trabajo.}
$$
