# Introducción

El individuo se comporta, es decir escoge cursos de acción cuando es llamado a hacerlo.

## Axiomas de Elección.

El problema de decisión se puede representar imaginando un conjunto $X$ que contiene el universo de acciones que eventualmente el individuo puede tener a su alcance. Una elección consiste en seleccionar una acción $x \in X$ entre todas las posibles.

Para representar formalmente este comportamiento, se plantean los siguientes axiomas:

### **Axioma 1: Completitud**


El individuo puede comparar cualquier par de alternativas:

$$
x \succeq x' \quad \lor \quad x' \succeq x \quad \forall x, x' \in X
$$

### **Axioma 2: Transitividad**

Si una opción es al menos tan buena como otra, y esta última es al menos tan buena como una tercera, entonces la primera es al menos tan buena como la tercera:

$$
x \succeq x', \, x' \succeq x'' \Rightarrow x \succeq x''
$$

### **Axioma 3: Continuidad**

Los conjuntos de alternativas al menos tan preferidas o menos preferidas que una opción dada deben ser cerrados:

$$
\forall x \in X, \quad \{ x' \in X : x' \succeq x \} \quad \text{y} \quad \{ x' \in X : x \succeq x' \} \quad \text{son cerrados}
$$


Esto significa que si una suceción de opciones ${X_n^{'}}$ es al menos tan buena como $x$, y converge a algún $x^{'}$ entonces $x^{'}$ tambien deber ser almenos tan bueno como $x$.

Cuando estos tres axiomas se cumplen es posible utilizar una función de utilidad para representar las preferencias del individuo mediante unafunción de utilidad.

### **Función de Utilidad**

Entonces existe una función de utildiad $u : X \rightarrow \mathbb{R}$ tal que asocia a cada acto $x$ un valor real $u(x)$, que representa la preferencia $\succeq$:

$$
x \succeq x' \Leftrightarrow u(x) \geq u(x') \space \forall x,x^{'} \in X
$$

### **Elección optima**

Asi existiendo esta función de utilidad, es posible representar el problema de elección como uno de maximización, donde el individuo selecciona la acción que le da mayor utilidad:

$$
x^*(X) = arg\max_{x \in X} u(x)
$$

En este sentido decimos que la utilida es **ordinal**, en contraposición a **cardinal**. Es ordinal por que nos interesa de ella es describir el comportamiento del individuo a partir del ordenamiento de las acciones que representa.


### Clasificación de Bienes

La utilidad marginal de cada bien determina si es considerado positivo, negativo o neutro.

- **Bien**: $ \frac{\partial u}{\partial x_\ell} > 0 $ Utilidad creciente.
- **Mal**: $\frac{\partial u}{\partial x_\ell} < 0 $ Utilidad decreciente.
- **Neutro**: $\frac{\partial u}{\partial x_\ell} = 0 $ Utilidad constante.


### Curvas de indiferencia


Una curva de indiferencia representa todas las combinaciones de bienes que otorgan al individuo el mismo nivel de utilidad.

$$
u(x_1, x_2) = \bar{u}
$$

Condicion de indiferencia 

$$
\frac{\partial u}{\partial x_1} dx_1 + \frac{\partial u}{\partial x_2} dx_2 = 0
$$

Despejando tenemos:


$$
\frac{dx_2}{dx_1} = -\frac{\frac{\partial u}{\partial x_1}}{\frac{\partial u}{\partial x_2}} = -\frac{u_1}{u_2}
$$
E valor absoluto de la solución representa la **tasa marginal de sustitución objetiva**, porque representa la maxima cantidad que el individuo esta dispuesto a sacrificar del bien 2 en sustitución de una unidad del bien 1.

# Primera Curva

Definimos las funcion de utilidad de la siguiente forma:

$$
U(x_1,x_2 )=hx_1+ox_2
$$

Donde $h \land o \geq 1 $ Por lo que sus curvas de indiferencia serian líneas rectas.

Por lo que si $h = 2 \land o = 4 \rightarrow U(x_1,x_2)=2x_1+4x_2$

La curva de indiferencia nos quedaria de la siguiente forma.

$$
x_2 = \frac{ U(x_1,x_2)}{4} - \frac{x_1}{2}
$$

Graficando con matplotlib.


```python
h = 2;o = 4

#definimos x_1
x1 = np.linspace(0,40,200)
#Niveles de utilidad que graficaremos
u_levels = [20,40,60]
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,5))

for u in u_levels:
    x2 = (u/o)- (h/o)*x1
    ax.plot(x1,x2,label = f'U = {u}')

ax.set_title('Curvas de indiferencia 1° función $U(x_1,x_2)=2x_1+4x_2$')
ax.set_ylabel('$x_2$');ax.set_xlabel('$x_1$')
ax.set_ylim(0,20);ax.set_xlim(0,40)
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```

# Óptimo del consumidor

Definimos las funcion de utilidad de la siguiente forma:

$$
U(x_1,x_2 )=hx_1+ox_2
$$

Donde $h \land o \geq 1 $ Por lo que sus curvas de indiferencia serian líneas rectas.

Por lo que si $h = 2 \land o = 4 \rightarrow U(x_1,x_2)=2x_1+4x_2$

La curva de indiferencia nos quedaria de la siguiente forma.

$$
x_2 = \frac{ U(x_1,x_2)}{4} - \frac{x_1}{2}
$$

La restricción Presupuestaria tiene la siguiente forma:

$$
110 = 52x_1 + 55x_2
$$

Para calcular el óptimo primero derivamos las **CPO**:

1. $\frac{\partial U}{\partial x_1}, \frac{\partial U}{\partial x_2}$
2. Como la función de utildiad es lineal, las curvas de indiferencia seran rectas con pendiente cosntante definida como la **TMS**:
    $$TMS = - \frac{\partial U / \partial x_1}{\partial U / \partial x_2} = -\frac{1}{2} $$
3. Relación de los precios: $$-\frac{p_{x_1}}{p_{x_2}} = -\frac{52}{55}$$
4. Comparamos con el modulo de la **TMS**.
    $|TMS| = \frac{1}{2} < \frac{52}{55}$ Lo que implica una solución de esquina. El consumidor consume solo $x_2$, ya qeu la pendiente de la restricción es más pronunciada que la curva de indiferencia, por lo que elindividuo destinara todo su ingreso a consumir unicamente $x_2$. Así se impone la siguiente solución.
5. $x_1^* = 0, \quad x_2^* = \frac{110}{55} = 2.$ El nivel de utildiad alcanzado es de $U(x_1^*, x_2^*) = 2(0) + 4(2) = 8$.
6. La curva de indiferencia con respecto a un nivel de utilida de 8 es de $$2x_1 + 4x_2 = 8 \quad \Rightarrow \quad x_2 = 2 - \frac{x_1}{2}.$$


```python
from sympy import symbols, Eq, solve
```

```python
h = 52;o = 55

#definimos x_1
x1 = np.linspace(0,20,200)
#Niveles de utilidad que graficaremos
u_levels = [20,30,40,50]
```

```python
x1_sym, x2_sym = symbols('x1 x2')
```

```python
eq1 = Eq(x2_sym, (110 - h * x1_sym)/o)  
eq2 = Eq(x2_sym, 2 - x1_sym / 2) 
```

```python
sol = solve((eq1, eq2), (x1_sym, x2_sym))
x_1_eq, x_2_eq = float(sol[x1_sym]), float(sol[x2_sym])
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,5))

presupuesto = (110/o) - (h*x1)/o
x2 = 2 - (x1/2)
ax.plot(x1,presupuesto,label = 'Restricción Presupuestaria', color = 'black')
ax.plot(x1,x2,label='Curva de indiferencia',color='green')
ax.plot(x_1_eq,x_2_eq, 'ko', label='Punto de equilibrio',color = "red")
ax.annotate('Pto equilibrio', xy=(x_1_eq, x_2_eq), xytext=(x_1_eq + 0.2, x_2_eq + 0.2),
             arrowprops=dict(arrowstyle='->',color='r'))
ax.set_title('Solución Optima para $U(x_1,x_2)=2x_1+4x_2$')
ax.set_ylabel('$x_2$');ax.set_xlabel('$x_1$')
ax.set_ylim(0,5);ax.set_xlim(0,5)
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```

La curva de indiferencia de Tita intersecta a la restriccion presupuestaria justo en un pto esquina, indicandonos que ella prefiere consumir solo $X_2 \text{(Irse en van)}$ le otorga mayor utilidad marginal relativa al costo.

Por lo que Tita destinara todo su presupuesto en arrendar una van con sus amigas para el paseo de curso.


# Óptimo del consumidor 2° Curva 

Definimos la función de utilidad de la siguiente forma:

$$
U(x_1,x_2 )= \sqrt{2}x_1 + x_2
$$

Donde ambos términos son positivos y lineales en su comportamiento, por lo que las curvas de indiferencia serán líneas rectas.

La restricción presupuestaria tiene la siguiente forma:

$$
110 = 52x_1 + 55x_2
$$

Para calcular el óptimo primero derivamos las **CPO**:

1. $\frac{\partial U}{\partial x_1} = \sqrt{2}, \quad \frac{\partial U}{\partial x_2} = 1 $

2. Como la función de utilidad es lineal, las curvas de indiferencia serán rectas con pendiente constante definida como la **TMS**:

$$
TMS = - \frac{\partial U / \partial x_1}{\partial U / \partial x_2} = -\sqrt{2}
$$

3. Relación de los precios:

$$
-\frac{p_{x_1}}{p_{x_2}} = -\frac{52}{55}
$$

4. Comparamos con el módulo de la **TMS**:

$$
|\text{TMS}| = \sqrt{2} \approx 1.41 > \frac{52}{55} \approx 0.945
$$

Esto implica una **solución de esquina**. El consumidor consume solo $x_1$, ya que valora relativamente más $x_1$ y su TMS es mayor que la pendiente de la recta presupuestaria. Por lo tanto, el individuo destinará todo su ingreso a consumir únicamente $x_1$. Así se impone la siguiente solución:

$$
x_1^* = \frac{110}{52} \approx 2.115, \quad x_2^* = 0
$$

El nivel de utilidad alcanzado es:

$$
U(x_1^*, x_2^*) = \sqrt{2} \cdot \frac{110}{52} \approx 1.414 \cdot 2.115 \approx 2.99
$$

La curva de indiferencia correspondiente a ese nivel de utilidad es:

$$
\sqrt{2}x_1 + x_2 = \sqrt{2} \cdot \frac{110}{52} \quad \Rightarrow \quad x_2 = \sqrt{2} \cdot \left(\frac{110}{52} - x_1\right)
$$


```python
h = 52;o = 55

#definimos x_1
x1 = np.linspace(0,20,200)
#Niveles de utilidad que graficaremos
u_levels = [20,30,40,50]
```

```python
x1_sym, x2_sym = symbols('x1 x2')
```

```python
eq1 = Eq(x2_sym, (110 - h * x1_sym)/o)  
eq2 = Eq(x2_sym, np.sqrt(2)*((110/52)-x1_sym)) 
```

```python
sol = solve((eq1, eq2), (x1_sym, x2_sym))
x_1_eq, x_2_eq = float(sol[x1_sym]), float(sol[x2_sym])
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,5))

presupuesto = (110/o) - (h*x1)/o
x2 = np.sqrt(2)*((110/52)-x1)
ax.plot(x1,presupuesto,label = 'Restricción Presupuestaria', color = 'black')
ax.plot(x1,x2,label='Curva de indiferencia',color='green')
ax.plot(x_1_eq,x_2_eq, 'ko', label='Punto de equilibrio',color = "red")
ax.annotate('Pto equilibrio', xy=(x_1_eq, x_2_eq), xytext=(x_1_eq + 0.3, x_2_eq + 0.4),
             arrowprops=dict(arrowstyle='->',color='r'))
ax.set_title('Solución Optima para $U(x_1,x_2 )= \sqrt{2}x_1 + x_2$')
ax.set_ylabel('$x_2$');ax.set_xlabel('$x_1$')
ax.set_ylim(0,5);ax.set_xlim(0,5)
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```

La curva de indiferencia de Tita intersecta a la restriccion presupuestaria justo en un pto esquina, indicandonos que ella prefiere consumir solo $X_1 \text{(Irse en Cabify)}$ le otorga mayor utilidad marginal relativa al costo.

Por lo que Tita destinara todo su presupuesto en solocitar un cabify para irse al aeropuerto.

U(x_1, x_2) = \ln(4)x_1 + \sqrt{2}x_2
$$

Donde ambos coeficientes son constantes positivas, por lo que las curvas de indiferencia son líneas rectas.

La restricción presupuestaria tiene la siguiente forma:

$$
110 = 52x_1 + 55x_2
$$

Para calcular el óptimo primero derivamos las **CPO**:

1. $$\frac{\partial U}{\partial x_1} = \ln(4) , \frac{\partial U}{\partial x_2} = \sqrt{2} $$

2. Como la función de utilidad es lineal, las curvas de indiferencia son rectas con pendiente constante definida como la **TMS**:

$$
TMS = -\frac{\frac{\partial U}{\partial x_1}}{\frac{\partial U}{\partial x_2}} = -\frac{\ln(4)}{\sqrt{2}} \approx -\frac{1.386}{1.414} \approx -0.98
$$

3. Relación de precios:

$$
-\frac{p_{x_1}}{p_{x_2}} = -\frac{52}{55} \approx -0.945
$$

4. Comparando con el módulo de la **TMS**:

$$
|TMS| \approx 0.98 > \frac{52}{55} \approx 0.945
$$

Esto implica una **solución de esquina**. El consumidor consume solamente $x_1$, ya que valora relativamente más $x_1$. Entonces el individuo destina todo su ingreso a consumir únicamente $x_1$:

$$
x_1^* = \frac{110}{52} \approx 2.115, \quad x_2^* = 0
$$

El nivel de utilidad alcanzado es:

$$
U(x_1^*, x_2^*) = \ln(4) \cdot \frac{110}{52} \approx 1.386 \cdot 2.115 \approx 2.93
$$

La curva de indiferencia correspondiente a ese nivel de utilidad es:

$$
\ln(4)x_1 + \sqrt{2}x_2 = 2.93 \quad \Rightarrow \quad 
x_2 = \frac{2.93 - \ln(4)x_1}{\sqrt{2}}
$$


```python
h = 52;o = 55

#definimos x_1
x1 = np.linspace(0,20,200)
#Niveles de utilidad que graficaremos
u_levels = [20,30,40,50]
```

```python
x1_sym, x2_sym = symbols('x1 x2')
```

```python
eq1 = Eq(x2_sym, (110 - h * x1_sym)/o)  
eq2 = Eq(x2_sym, (2.93 - np.log(4)*x1_sym)/np.sqrt(2)) 
```

```python
sol = solve((eq1, eq2), (x1_sym, x2_sym))
x_1_eq, x_2_eq = float(sol[x1_sym]), float(sol[x2_sym])
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,5))

presupuesto = (110/o) - (h*x1)/o
x2 = (2.93 - np.log(4)*x1)/np.sqrt(2)
ax.plot(x1,presupuesto,label = 'Restricción Presupuestaria', color = 'black')
ax.plot(x1,x2,label='Curva de indiferencia',color='green')
ax.plot(x_1_eq,x_2_eq, 'ko', label='Punto de equilibrio',color = "red")
ax.annotate('Pto equilibrio', xy=(x_1_eq, x_2_eq), xytext=(x_1_eq + 0.3, x_2_eq + 0.4),
             arrowprops=dict(arrowstyle='->',color='r'))
ax.set_title('Solución Optima para $U(x_1, x_2) = \ln(4)x_1 + \sqrt{2}x_2$')
ax.set_ylabel('$x_2$');ax.set_xlabel('$x_1$')
ax.set_ylim(0,5);ax.set_xlim(0,5)
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```

Por lo que Tita destinara todo su presupuesto en solocitar un cabify para irse al aeropuerto.

# Definimos la función de utilidad de la siguiente forma:

$$
U(x_1, x_2) = \left( x_1^{\rho} + x_2^{\rho} \right)^{\frac{1}{\rho}} \quad \text{donde } \rho = \frac{\theta - 1}{\theta}, \ \theta > 1
$$

Esta es una función de utilidad **CES** (Elasticidad Constante de Sustitución), la cual permite representar distintos grados de sustitución entre los bienes según el parámetro $\theta$. Sus curvas de indiferencia son convexas (no lineales) y su forma depende del valor de $\theta$.

Por ejemplo, si $\theta = 2$, entonces:

$$
\rho = \frac{2 - 1}{2} = \frac{1}{2}
$$

La función de utilidad se convierte en:

$$
U(x_1, x_2) = \left( x_1^{1/2} + x_2^{1/2} \right)^2
$$

Para graficar las curvas de indiferencia, se iguala la utilidad a un nivel fijo $\bar{U}$, y se despeja $x_2$ como función de $x_1$:

Partimos de:

$$
U(x_1, x_2) = \bar{U} \Rightarrow \left( x_1^{\rho} + x_2^{\rho} \right)^{\frac{1}{\rho}} = \bar{U}
$$

Elevamos ambos lados a $\rho$:

$$
x_1^{\rho} + x_2^{\rho} = \bar{U}^{\rho}
$$

Despejamos $x_2$:

$$
x_2 = \left( \bar{U}^{\rho} - x_1^{\rho} \right)^{\frac{1}{\rho}}
$$


Entonces si:

1. $\theta \rightarrow 1$: La función tiene forma de Cobb-Douglas.
2. $\theta \rightarrow \infty$: Los bienes se vuelven sustitutos perfectos. 
3. $\theta \rightarrow 0$: los bienes se comportan como colpementarios perfectos.

Graficando con matplotlib.


```python
#Prueben modificando o generando una lsita con distintos valores de theta para ver que curvas le dan!
theta = 2
rho = (theta - 1) / theta
inv_rho = theta / (theta - 1)

x_1 = np.linspace(0, 10, 200)
u_levels = [2, 4, 6]
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

for U in u_levels:
    x_2 = (U**(1/inv_rho) - x_1**rho)**(1/rho)
    x_2 = np.where(np.isreal(x_2), x_2.real, np.nan) 
    ax.plot(x_1, x_2, label=f'U = {U}')

ax.set_title('Curvas de indiferencia para CES')
ax.set_ylabel('$X_2$');ax.set_xlabel('$X_1$')
ax.set_ylim(0,10);ax.set_xlim(0,10)
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```

Las curvas de indiferencia nos muestran que el consumidor está dispuesto a sustituir un bien por otro, pero **con una tasa de sustitución decreciente**. A diferencia de las preferencias cuasilineales, aquí **ambos bienes son valorados de manera creciente**, y su grado de sustitución depende directamente del parámetro $\theta$:

- Cuando $\theta$ es bajo (pero mayor que 1), las curvas son más curvas (complementarios).
- Cuando $\theta$ es alto, las curvas se aproximan a líneas rectas (sustitutos).

Esto significa que para un consumidor con esta función de utilidad, **ambos bienes son necesarios para alcanzar un mayor nivel de satisfacción**, pero la forma exacta en que los combina depende del valor de $\theta$. No hay una preferencia fija por un solo bien como en el caso cuasilineal; más bien, **la elección óptima dependerá del equilibrio entre precios y utilidad marginal relativa**.

# Demandas Marshallianas

## Supuesto

La función de utilidad del consumidor es CES:

$$
U(x_1, x_2) = \left(x_1^{\frac{\theta - 1}{\theta}} + x_2^{\frac{\theta - 1}{\theta}}\right)^{\frac{\theta}{\theta - 1}}, \quad \theta > 1
$$

Y su restricción presupuestaria está dada por:

$$
p_1 x_1 + p_2 x_2 = M
$$


## Formulamos el Lagrangeano

Planteamos el problema de maximización:

$$
\mathcal{L}(x_1, x_2, \lambda) = \left(x_1^{\frac{\theta - 1}{\theta}} + x_2^{\frac{\theta - 1}{\theta}}\right)^{\frac{\theta}{\theta - 1}} + \lambda (M - p_1 x_1 - p_2 x_2)
$$

## Derivadas parciales (CPO)
$
Derivamos con respecto a $x_1$, $x_2$, y $\lambda$:

$$
\frac{\partial \mathcal{L}}{\partial x_1} = x_1^{\frac{\theta - 1}{\theta} - 1} \left(x_1^{\frac{\theta - 1}{\theta}} + x_2^{\frac{\theta - 1}{\theta}}\right)^{\frac{1}{\theta - 1}} - \lambda p_1 = 0
$$

$$
\frac{\partial \mathcal{L}}{\partial x_2} = x_2^{\frac{\theta - 1}{\theta} - 1} \left(x_1^{\frac{\theta - 1}{\theta}} + x_2^{\frac{\theta - 1}{\theta}}\right)^{\frac{1}{\theta - 1}} - \lambda p_2 = 0
$$

$$
\frac{\partial \mathcal{L}}{\partial \lambda} = M - p_1 x_1 - p_2 x_2 = 0
$$


## Igualamos ecuaciones 1 y 2 para eliminar $\lambda$

De las dos primeras condiciones:

$$
\frac{x_1^{\frac{\theta - 1}{\theta} - 1}}{p_1} = \frac{x_2^{\frac{\theta - 1}{\theta} - 1}}{p_2}
$$

Sabemos que:

$$
\frac{\theta - 1}{\theta} - 1 = -\frac{1}{\theta}
$$

Entonces:

$$
\frac{1}{x_1^{1/\theta}} = \frac{p_1}{p_2} \cdot \frac{1}{x_2^{1/\theta}} \Rightarrow p_2 x_2^{1/\theta} = p_1 x_1^{1/\theta}
$$


## Despejamos $x_1$ y $x_2$

A partir de la ecuación anterior:

$$
x_1 = \left( \frac{p_2 x_2^{1/\theta}}{p_1} \right)^{\theta}
$$

$$
x_2 = \left( \frac{p_1 x_1^{1/\theta}}{p_2} \right)^{\theta}
$$


## Sustituimos en la restricción presupuestaria

Sustituimos $x_1$ en la restricción:

$$
p_1 x_1 + p_2 x_2 = M
$$

$$
p_1 \left( \frac{p_2 x_2^{1/\theta}}{p_1} \right)^{\theta} + p_2 x_2 = M
$$

$$
\frac{p_2^{\theta}}{p_1^{\theta - 1}} x_2 + p_2 x_2 = M
$$

$$
x_2 \left( \frac{p_2^{\theta}}{p_1^{\theta - 1}} + p_2 \right) = M
$$

Entonces:

$$
x_2^*(p, m) = \frac{M}{\frac{p_2^{\theta}}{p_1^{\theta - 1}} + p_2}
$$

Y análogamente:

$$
x_1^*(p, m) = \frac{M}{\frac{p_1^{\theta}}{p_2^{\theta - 1}} + p_1}
$$


## Optimos: Demandas Marshallianas

Las demandas Marshallianas son:

$$
x_1^*(p, m) = \frac{M}{\frac{p_2 p_1^{\theta}}{p_2^{\theta}} + p_1}
$$

$$
x_2^*(p, m) = \frac{M}{\frac{p_1 p_2^{\theta}}{p_1^{\theta}} + p_2}
$$


Por lo tanto, el ingreso M se distribuye entre ambos bienes, por lo que individuo ajusta su consumo de cada bien en respuesta de los precios relativos y al ingreso, teniendo en cuenta la facilidad o dificultad de sustituir un bien por otro controlado por el parametro $\theta$.

Cuanto mayor sea $\theta$, más sensible sera el consumidor al precio relativo. Cuanto menor sea $\theta$, mas estaticas seran las proporciones de consumo entre los bienes.

```python
thetas = [0.2,1.2,10]
m_s = [5,10,20]

x_1 = np.linspace(0, 20, 200)
p_1,p_2 = 1,1
```

```python
for theta in thetas:
    rho = (theta - 1) / theta
    inv_rho = theta / (theta - 1)

    fig, ax = plt.subplots(figsize=(8, 5))
    
    u_levels = [2, 5, 15]
    for U in u_levels:
        x_2 = (U**(1/inv_rho) - x_1**rho)**(1/rho)
        x_2 = np.where(np.isreal(x_2), x_2.real, np.nan) 
        ax.plot(x_1, x_2, label=f'U = {U}', linestyle='-')

    for m in m_s:
        x2_budget = (m - p_1 * x_1) / p_2
        ax.plot(x_1, x2_budget, linestyle='--', label=f'm = {m}', alpha=0.7)

    ax.set_title(f'Curvas CES con restricciones presupuestarias (θ = {theta})')
    ax.set_ylabel('$X_2$');ax.set_xlabel('$X_1$')
    ax.set_xlim(0, 20)
    ax.set_ylim(0, 20)
    ax.grid(True)
    ax.legend()
    plt.tight_layout()
    plt.show()
```
